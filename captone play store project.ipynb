```python
# Missing Values/Null Values Count
df = df.dropna()
```

```python
df['Installs'] = df['Installs'].str.replace("+", "")
```

```python
# prompt: change data type of Installs colum

df['Installs'] = df['Installs'].str.replace(',', '').astype(int)
```

```python
df['Rating'] = df['Rating'].astype(int)
df['Reviews'] = df['Reviews'].astype(int)
```

#### Missing Values/Null Values

```python
# Visualizing the missing value
pd.isnull(df).sum()
```

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# prompt: total installs by category

cat_installs = df.groupby('Category')['Installs'].sum()/1000000
print(cat_installs)
```

```python
# prompt: draw pia chart category vs installs

import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))
plt.pie(cat_installs, labels=cat_installs.index, autopct="%1.1f%%", startangle=90)
plt.title("Category vs Installs")
plt.show()
```

```python
# prompt: category count apps

df['Category'].value_counts()
```

```python
# prompt: draw pia chart category vs app

import matplotlib.pyplot as plt
cat_apps = df.groupby('Category')['App'].count()
plt.figure(figsize=(15,8))
plt.pie(cat_apps, labels=cat_apps.index, autopct="%1.1f%%", startangle=90)
plt.title("Category vs Apps")
plt.show()
```

```python
# prompt: bottom app that installs

df.groupby('App')['Installs'].sum().sort_values(ascending=True).head(10)
```

```python
# prompt: top five category

df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head()
```
